```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Use tinyllama (open-access, no key needed)
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# Create text-generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# Example medical question
prompt = "You are a helpful medical assistant. A patient says they have chest pain and shortness of breath. What questions should I ask to better understand their condition?"

# Generate response
response = generator(
    prompt,
    max_length=300,
    do_sample=True,
    top_p=0.9,
    temperature=0.7
)

print(response[0]["generated_text"])
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["quit", "exit", "stop"]:
        print("Chatbot: Goodbye! Take care.")
        break

    response = generator(
        f"You are a helpful medical assistant. Patient says: {user_input}\nAssistant:",
        max_length=250,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        truncation=True
    )

    print("Chatbot:", response[0]["generated_text"])
```
